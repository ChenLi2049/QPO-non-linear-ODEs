```python
#MHR12CJ0JmJ
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

Fs=100#
dt=1/Fs#
n=200000#

x=np.zeros(n,dtype=float)
y=np.zeros(n,dtype=float)
vx=np.zeros(n,dtype=float)
vy=np.zeros(n,dtype=float)
t=np.arange(n)*dt

=1
1=3
2=3

C=4.5
J0=-2.2
Jm=2.2
m=1.5
l=1
nT=62500#J
nt1=16000#J
nt2=26000#J

J_S=np.zeros(n,dtype=float)
J=np.zeros(n,dtype=float)
for i in range(0,nT):
    J_S[i]=Jm*(1-(1-i/nT))**m+J0
    J[i]=J_S[i]+C*(np.random.random()-0.5)
for i in range(nt1,nt2):
    J_S[i]=Jm*((nt2-i)/(12000))**l-1.6
    J[i]=J_S[i]+C*(np.random.random()-0.5)
for i in range(nT,n):
    J_S[i]=J_S[i-nT]
    J[i]=J_S[i]+C*(np.random.random()-0.5)

#
for i in range(0,n-1):
    vx[i+1]=-*(x[i]**3)+1*(x[i]**2)+y[i]+J[i]
    vy[i+1]=-2*(x[i]**2)-y[i]
    x[i+1]=x[i]+vx[i+1]*dt
    y[i+1]=y[i]+vy[i+1]*dt

#
fig=plt.figure(figsize=(10,20))

#xpercent
percent=0.7
ax1=fig.add_subplot(4,1,1)
ax1.plot(t[:int(percent*n)],x[:int(percent*n)],'r',lw=1,label=r'$\beta_1=$'+str(1)+' , '+r'$\beta_2=$'+str(2)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+' , '+r'$J_m=$'+str(Jm))
ax1.set_title(r'Light Curve',fontstyle='italic')
ax1.set_xlabel(r'Time (s)')
ax1.set_ylabel(r'Relative Intensity')
ax1.set_xlim(0, percent*n*dt)
ax1.legend(loc = 'upper right')
ax1.grid()

#Jpercent
ax2=fig.add_subplot(4,1,2)
ax2.plot(t[:int(percent*n)],J[:int(percent*n)],'g',lw=1,label=r'$J$')
ax2.plot(t[:int(percent*n)],J_S[:int(percent*n)],'y',lw=1,label=r'$J_S$')
ax2.plot(t[:int(percent*n)],np.ones(int(percent*n))*np.mean(J_S),'#06470c',lw=1,label=r'$<J>=$'+str(np.mean(J)))
ax2.set_title(r'Input Function J',fontstyle='italic')
ax2.set_xlabel(r'Time (s)')
ax2.set_ylabel(r'J')
ax2.set_xlim(0, percent*n*dt)
ax2.legend(loc = 'upper right')
ax2.grid()

#x-y
ax3=fig.add_subplot(4,1,3)
x_t=np.linspace(-1.5,1,100)
y1=*(x_t**3)-1*(x_t**2)-J0
y2=-2*(x_t**2)
ax3.plot(x_t,y1,'k',label=r'$y=x^3-\beta_1 x^2-J_0$')
ax3.plot(x_t,y2,'k,--',label=r'$y=-\beta_2 x^2$')
persent_xy=0.2
ax3.plot(x[:int(persent_xy*n)],y[:int(persent_xy*n)],'r',lw=0.3,label=r'$\beta_1=$'+str(1)+' , '+r'$\beta_2=$'+str(2)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+' , '+r'$J_m=$'+str(Jm))
ax3.set_title(r'x-y',fontstyle='italic')
ax3.set_xlabel(r'x')
ax3.set_ylabel(r'y')
ax3.legend(loc = 'upper right')
ax3.grid()

#x,
ax4=fig.add_subplot(4,1,4)
ax4.plot(np.arange(int(n/2))*Fs/n,2*np.abs(fft(x))[:int(n/2)]/n,'r',lw=1,label=r'$\beta_1=$'+str(1)+' , '+r'$\beta_2=$'+str(2)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+' , '+r'$J_m=$'+str(Jm)) 
ax4.set_title(r'Power Density Spectrum',fontstyle='italic')
ax4.set_xlabel(r'f (Hz)')
ax4.set_ylabel(r'Relative Power')
ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.legend(loc = 'upper right')
ax4.grid()

plt.show()
```
