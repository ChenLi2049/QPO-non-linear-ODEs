```python
#MHRCJ0J

import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

Fs=100#
dt=1/Fs#
t_total=480
n=int(t_total/dt)#

x=np.zeros(n,dtype=float)
y=np.zeros(n,dtype=float)
vx=np.zeros(n,dtype=float)
vy=np.zeros(n,dtype=float)
t=np.arange(n)*dt

=1
1=3
2=1

C=4
J0=6.5
J=[]
#J=J0+C*(np.random.random(n)-0.5)#J
t1=5
t2=10
for j in range(0,int(t_total/t2)):
    for i in range(j*int(t2/dt),j*int(t2/dt)+int(t1/dt)):
        J.append(J0+C*(np.random.random()-0.5))
    for i in range(j*int(t2/dt)+int(t1/dt),(j+1)*int(t2/dt)):
        J.append(J0)


#
for i in range(0,n-1):
    vx[i+1]=-*(x[i]**3)+1*(x[i]**2)+y[i]+J[i]
    vy[i+1]=-2*(x[i]**2)-y[i]
    x[i+1]=x[i]+vx[i+1]*dt
    y[i+1]=y[i]+vy[i+1]*dt

#
x_wn=np.zeros(n,dtype=float)
y_wn=np.zeros(n,dtype=float)
vx_wn=np.zeros(n,dtype=float)
vy_wn=np.zeros(n,dtype=float)
for i in range(0,n-1):
    vx_wn[i+1]=-*(x_wn[i]**3)+1*(x_wn[i]**2)+y_wn[i]+J0
    vy_wn[i+1]=-2*(x_wn[i]**2)-y_wn[i]
    x_wn[i+1]=x_wn[i]+vx_wn[i+1]*dt
    y_wn[i+1]=y_wn[i]+vy_wn[i+1]*dt

#
fig=plt.figure(figsize=(10,20))

#xpercent
percent=0.2
ax1=fig.add_subplot(4,1,1)
ax1.plot(t[:int(percent*n)],x[:int(percent*n)],'r',lw=1,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0))
ax1.plot(t[:int(percent*n)],x_wn[:int(percent*n)],'m',lw=1,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+r'(without noise)')
ax1.set_title(r'Light Curve',fontstyle='italic')
ax1.set_xlabel(r'Time (s)')
ax1.set_ylabel(r'Relative Intensity')
ax1.set_xlim(0, percent*n*dt)
ax1.legend(loc = 'upper right')
ax1.grid()

#Jpercent
ax2=fig.add_subplot(4,1,2)
ax2.plot(t[:int(percent*n)],J[:int(percent*n)],'g',lw=1,label=r'$J$')
ax2.plot(t[:int(percent*n)],np.ones(int(percent*n))*J0,'y',lw=1,label=r'$J_0=$'+str(J0))
ax2.set_title(r'Input Function J',fontstyle='italic')
ax2.set_xlabel(r'Time (s)')
ax2.set_ylabel(r'J')
ax2.set_xlim(0, percent*n*dt)
ax2.legend(loc = 'upper right')
ax2.grid()

#x-y
ax3=fig.add_subplot(4,1,3)
x_t=np.linspace(0,3.5,100)
y1=*(x_t**3)-1*(x_t**2)-J0
y2=-2*(x_t**2)
ax3.plot(x_t,y1,'k',label=r'$y=x^3-\beta_1 x^2-J_0$')
ax3.plot(x_t,y2,'k,--',label=r'$y=-\beta_2 x^2$')
persent_xy=0.1
ax3.plot(x[:int(persent_xy*n)],y[:int(persent_xy*n)],'r',lw=0.5,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0))
ax3.plot(x_wn[:int(persent_xy*n)],y_wn[:int(persent_xy*n)],'m',lw=0.5,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+r'(without noise)')
ax3.set_title(r'x-y(Nullclines & Simulations)',fontstyle='italic')
ax3.set_xlabel(r'x')
ax3.set_ylabel(r'y')
ax3.legend(loc = 'upper right')
ax3.grid()

#x,
ax4=fig.add_subplot(4,1,4)
ax4.plot(np.arange(int(n/2))*Fs/n,2*np.abs(fft(x))[:int(n/2)]/n,'r',lw=1,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)) 
ax4.plot(np.arange(int(n/2))*Fs/n,2*np.abs(fft(x_wn))[:int(n/2)]/n,'m',lw=1,label=r'$\beta_1=\beta_2=$'+str(1)+' , '+r'$C=$'+str(C)+' , '+r'$J_0=$'+str(J0)+r'(without noise)') 
ax4.set_title(r'Power Density Spectrum',fontstyle='italic')
ax4.set_xlabel(r'f (Hz)')
ax4.set_ylabel(r'Relative Power')
ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.legend(loc = 'upper right')
ax4.grid()

plt.show()
```
